```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("mw_govjob_listings_scraped.csv")
```

```python
df
```

```python
df = df.drop('Description', axis=1)
df = df.drop('Link', axis=1)
```

```python
df.drop(df.loc[df['Salary']==''].index, inplace=True)
```

```python
df['Salary'] = df['Salary'].str.replace('Starting at','')
```

```python
df['Salary'] =df['Salary'] .str[:8]
```

```python
df['Salary'] = df['Salary'].str.replace('$','')
df['Salary'] = df['Salary'].str.replace('NF','')
df['Salary'] = df['Salary'].str.replace('(','')
df['Salary'] = df['Salary'].str.replace(',','')
df['Salary'] = df['Salary'].str.replace('0 GS','')
df['Salary'] = df['Salary'].str.replace('0 HS','')
```

```python
df.dtypes
```

```python
df['Salary'] = df['Salary'].replace(np.nan, 0)
```

```python
df['Salary'] = df['Salary'].str.strip()
```

```python
df['Salary'] = df['Salary'].fillna('')
```

```python
df['Salary'] = df['Salary'].str.strip()
```

```python
# df["Salary"] = df['Salary'].str.replace(' ', '').astype(float)
```

```python
df['Salary'] = pd.to_numeric(df['Salary'], downcast='float')
```

```python
df['Salary'] = df['Salary'].apply(lambda x: x*2000 if x<100 else x)
```
